```python
df = ords_prods_merged[:1000000]
```

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Create new column
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check values
df['prices'].value_counts()
```

## If-statements with for-loops

```python
# Create new column
ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Set Loop
result = []

for value in ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Add result to dataframe
ords_prods_merged['busiest_day'] = result
```

```python
# Check Values
ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

# Step 2

```python
# Set New Loop
result_a = []

for value in ords_prods_merged["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_a.append("Busiest day")
  elif value == 4 or value == 3:
    result_a.append("Least busy")
  else:
    result_a.append("Regularly busy")
```

```python
# Add result_a to dataframe
ords_prods_merged['busiest_days'] = result_a
```

# Step 4 ID busiest hours of the day

```python
# Check hours of df
ords_prods_merged['order_hour_of_day'].value_counts()
```

##### Most orders: 10, 11, 14, 15, 13, 12, 16, and 9
##### Average orders: 17, 8, 18, 19, 20, 7, 21, and 22
##### Fewest orders: 23, 6, 0, 1, 5, 2, 4, and 3

```python
# Create for-loop with hours
busiest_periods = []

for value in ords_prods_merged["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    busiest_periods.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    busiest_periods.append("Fewest orders")
  else:
    busiest_periods.append("Average orders")
```

```python
# Apply to df
ords_prods_merged['busiest_period_of_day'] = busiest_periods
```
