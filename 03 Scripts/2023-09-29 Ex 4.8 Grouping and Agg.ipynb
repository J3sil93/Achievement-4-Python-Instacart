## Step 6

### Create a spending flag for each user based on the average price across all their orders


```python
# Split the data into groups based on the 'user_id' column.
# Apply the transform() function on the 'prices' to get the avg price
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] >=10, 'spender_type'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] <10, 'spender_type'] = 'Low spender'
```

```python
ords_prods_merge['spender_type'].value_counts(dropna = False)
```

## Step 7

### Determine frequent versus non-frequent customers.

```python
# Split the data into groups based on the 'user_id' column.
# Apply the transform() function on the 'days_since_last_order' to get the median
ords_prods_merge['past_orders_median'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['past_orders_median'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['past_orders_median'] <= 20) & (ords_prods_merge['past_orders_median'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['past_orders_median'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Printing the frequency
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Checking multiple columns
ords_prods_merge[['user_id', 'past_orders_median', 'frequency_flag']].head(20)
```
